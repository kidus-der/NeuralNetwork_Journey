```python
import math
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def f(x):
    return 3*x**2 - 2*x - 5
```

A straight forward function f(x) that takes an input and returns a resulting output of the same type.

```python
f(4.0)
```

```python
xs = np.arange(-5,5, 0.25)
print(xs)
```

An array from the starting integer to the ending integer on an interval, all defined using '.arange'

```python
ys = f(xs)
print(ys)
```

Our function f(x), applied on each element of the array xs

```python
plt.plot(xs, ys)
```

We can verify the above parabola is true, since with our input of 4.0, we received the output 35 and the graph above matches our results. Now

Now, how do we find the derivative of this function f(x) at each point x?

```python
h = 0.000001
x = (7/2)
print((f(x+h) - f(x))/h)
```
